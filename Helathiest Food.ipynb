# Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv('Top 100 Healthiest Food in the World.csv', index_col='Food')
data.head()
```

## Selecting & Exploration

```python
data = data[['Calories','Protein (g)','Protein (g)','Vitamin C (mg)','Antioxidant Score']]
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Misalnya 'data' sudah hanya berisi angka (tanpa kolom negara)
scaled_data = pd.DataFrame(
    StandardScaler().fit_transform(data),
    index=data.index,
    columns=data.columns
)
```

## Heatmaps & Dendrogram

```python
sns.clustermap(scaled_data, cmap='rainbow' , figsize=(20, 20), xticklabels=True, yticklabels=True)
```
